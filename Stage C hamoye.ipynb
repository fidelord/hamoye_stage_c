```python
import pandas as pd
```

```python
df= pd.read_csv('Data_for_UCI_named.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['stab'], axis = 1, inplace = True)
```

```python
df['Stabf']=1
```

```python
df.loc[df['stabf']=='unstable','Stabf']=0
```

```python
df.drop(['stabf'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = 'Stabf')
y = df[['Stabf']]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.2 , random_state= 1 )
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
normalised_train_df = scaler.fit_transform(x_train)
normalised_test_df = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn import *
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification
import numpy as np
from sklearn.metrics import accuracy_score
clf = ExtraTreesClassifier(random_state=1)
clf.fit(normalised_train_df, y_train)
accuracy = accuracy_score(y_true=y_test, y_pred=clf.predict(normalised_test_df))
print( 'Accuracy: {}' .format(accuracy)) #prints 53.0

```

```python
#question 2
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
model = ExtraTreesClassifier(random_state=1)
param_grid = {'n_estimators': [ 100,  200,  300,  400,  500,  600,  700,  800,  900, 1000, 1100,
        1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900],
 'max_features': ['auto', 'sqrt', 'log2'],
 'max_depth': [10, 20, 30, 40, 50, 60, 70, 80, 90, None],
 'min_samples_split': [2, 4, 6, 8],
 'min_samples_leaf': [1, 2, 4],
 'bootstrap': [True, False]}
randm_src = RandomizedSearchCV(estimator=model, param_distributions = param_grid,
                               cv = 2, n_iter = 10, n_jobs=-1)
randm_src.fit(normalised_train_df, y_train)
accuracy = accuracy_score(y_true=y_test, y_pred=randm_src.predict(normalised_test_df))
print( 'Accuracy: {}' .format(accuracy)) #prints 53.0

```

```python
#question 12
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1)
clf.fit(normalised_train_df, y_train)
accuracy = accuracy_score(y_true=y_test, y_pred=clf.predict(normalised_test_df))
print( 'Accuracy: {}' .format(accuracy)) #prints 53.0

```

```python
#question 13
import xgboost as xgb
model = xgb.XGBClassifier(random_state=1)
model.fit(normalised_train_df, y_train)
accuracy = accuracy_score(y_true=y_test, y_pred=model.predict(normalised_test_df))
print( 'Accuracy: {}' .format(accuracy)) 
```
